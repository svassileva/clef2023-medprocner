```python
df_train_split = pd.read_csv('train_split/train_split_80.tsv', sep='\t', header=0)
df_val_split = pd.read_csv('train_split/test_split_20.tsv', sep='\t', header=0)
```

```python
df_train_split.head()
```

```python
from ast import literal_eval
df_train_split['mentions'] = df_train_split['mentions'].apply(literal_eval)
df_train_split['tokens'] = df_train_split['tokens'].apply(literal_eval)
```

```python
train_mentions = dict()

for index, row in df_train_split.iterrows():
    if len(row['mentions']) > 0:
        for mention in row['mentions']:
            mention_text = mention['original_text'].lower()
            if not mention_text in train_mentions.keys():
                train_mentions[mention_text] = mention['code']
```

```python
len(train_mentions)
```

```python
df_val_split.head()
```

```python
df_val_split['mentions'] = df_val_split['mentions'].apply(literal_eval)
df_val_split['tokens'] = df_val_split['tokens'].apply(literal_eval)
```

```python
val_mentions = dict()

for index, row in df_val_split.iterrows():
    if len(row['mentions']) > 0:
        for mention in row['mentions']:
            mention_text = mention['original_text'].lower()
            if not mention_text in val_mentions.keys():
                val_mentions[mention_text] = mention['code']
```

```python
remove_items = set()
for item in train_mentions.keys():
    if item in val_mentions.keys():
        remove_items.add(item)
```

```python
len(remove_items)
```

```python
for remove_item in remove_items:
    train_mentions.pop(remove_item)
```

```python
def get_sentence_idx(filename):
    start_idx = filename.index('-b-')
    return int(filename[start_idx+3:])-1
```

```python
def get_original_filename(filename):
    start_idx = filename.index('-b-')
    return filename[0:start_idx]
```
